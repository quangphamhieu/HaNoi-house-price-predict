```python
import os
import pandas as pd

#Dữ liệu từ các quận crawl thành nhiều file dùng để gộp lại thành 1 file chung, giả sử làm với quận thanh xuân

# Đường dẫn tới thư mục chứa các file Excel
folder_path = 'crawl data/ThanhXuan'

# Lấy danh sách tất cả các file Excel trong thư mục
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Khởi tạo danh sách để lưu các dataframe
df_list = []

# Đọc từng file Excel và thêm vào danh sách
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    df_list.append(df)

# Gộp tất cả các dataframe lại thành một
merged_df = pd.concat(df_list, ignore_index=True)

# Ghi dataframe đã gộp vào file Excel mới
merged_df.to_excel('crawl data/ThanhXuan.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
# Đọc dữ liệu từ tập tin Excel
df = pd.read_excel("crawl data/ThanhXuan.xlsx")

# Hiển thị vài dòng đầu tiên của DataFrame
df.head()
```

```python
num_rows, num_cols = df.shape
print("Num rows: ", num_rows)
print("Num cols: ", num_cols)
```

```python
# Chọn các cột tiêu đề và giá tiền để kiểm tra trùng lặp (giả sử tên cột là 'Title' và 'Price's
sum(df.duplicated(subset=['Tiêu đề', 'Giá tiền']))
```

```python
df_unique = df.drop_duplicates(subset=['Tiêu đề', 'Giá tiền'])

# Ghi kết quả sau khi loại bỏ trùng lặp vào file Excel mới
df_unique.to_excel('crawl data/tx.xlsx', index=False)
```

```python
# Sử dụng Regex để tách thông tin phường
df['Phường'] = df['Địa chỉ'].str.extract(r'Phường\s+([^\d,]+)')

# Hiển thị DataFrame sau khi thêm cột Phường
df.head()
```

```python
df['Quận'] = 'Thanh Xuân'

# Hiển thị DataFrame sau khi thêm cột Quận
df.head()
```

```python
# Loại bỏ chữ "phòng" để giữ lại số
df['Số phòng ngủ'] = df['Số phòng ngủ'].str.replace(' phòng', '')

# Loại bỏ chữ "tầng" để giữ lại số
df['Số tầng'] = df['Số tầng'].str.replace(' tầng', '')

df['Số toilet'] = df['Số toilet'].str.replace(' phòng', '')
```

```python
# Chuyển đổi 'not found' thành NaN
df.replace('Not found', np.nan, inplace=True)

# Chuyển đổi các giá trị còn lại thành kiểu số
df['Số tầng'] = pd.to_numeric(df['Số tầng'], errors='coerce')
df['Số phòng ngủ'] = pd.to_numeric(df['Số phòng ngủ'], errors='coerce')
df['Số toilet'] = pd.to_numeric(df['Số toilet'], errors='coerce')
# Xóa các hàng nếu cả 'Số tầng' và 'Số phòng ngủ' đều là NaN
df = df.dropna(subset=['Số tầng', 'Số phòng ngủ'], how='all')

# Cập nhật các giá trị 'Số tầng' khi nó là NaN
df.loc[df['Số tầng'].isna(), 'Số tầng'] = df['Số phòng ngủ'].apply(lambda x: x - 1 if x > 1 else 1)

# Cập nhật các giá trị 'Số phòng ngủ' khi nó là NaN
df.loc[df['Số phòng ngủ'].isna(), 'Số phòng ngủ'] = df['Số tầng'] + 1

df.head()
```

```python
df['Số phòng ngủ'] = df['Số phòng ngủ'].astype(int)
df['Số tầng'] = df['Số tầng'].astype(int)
df['Số toilet'] = df['Số tầng'].astype(int)
```

```python
# Kiểm tra các giá trị trong cột "Số phòng"
print(df['Số toilet'].unique())
```

```python
# Loại bỏ chữ để giữ lại số
df["Diện tích"]=df["Diện tích"].str.replace('m²', '')
df['Đường vào'] = df['Đường vào'].str.replace('m', '')
df['Mặt tiền'] = df['Mặt tiền'].str.replace('m', '')
```

```python
df['Đường vào'] = pd.to_numeric(df['Đường vào'], errors='coerce')
df['Mặt tiền'] = pd.to_numeric(df['Mặt tiền'], errors='coerce')
```

```python
# Đảm bảo cột 'Diện tích' là kiểu số, loại bỏ đơn vị nếu cần
df['Đường vào'] = df['Đường vào'].astype(str).str.replace(',', '.').astype(float)
# Đảm bảo cột 'Diện tích' là kiểu số, loại bỏ đơn vị nếu cần
df['Mặt tiền'] = df['Mặt tiền'].astype(str).str.replace(',', '.').astype(float)
```

```python
# Kiểm tra các giá trị trong cột "Giá tiền"
print(df['Giá tiền'].unique())

# Loại bỏ các dòng có giá tiền là "Thỏa thuận"
df = df[df['Giá tiền'] != 'Thỏa thuận']
```

```python
num_rows, num_cols = df.shape
print("Num rows: ", num_rows)
print("Num cols: ", num_cols)
```

```python
# Loại bỏ các dòng có giá trị 'tỷ/m²' trong cột 'Giá tiền'
df = df[~df['Giá tiền'].str.contains('tỷ/m²', na=False)]

# Xử lý giá trị 'tỷ'
df['Giá tiền'] = df['Giá tiền'].str.replace(' tỷ', '', regex=False)

# Đảm bảo cột 'Diện tích' là kiểu số, loại bỏ đơn vị nếu cần
df['Diện tích'] = df['Diện tích'].astype(str).str.replace(',', '.').astype(float)
```

```python
# Xử lý các giá trị còn lại
def convert_price(row):
    price_str = row['Giá tiền']
    area = row['Diện tích']  # Giả sử cột diện tích đã là số
    if 'triệu/m²' in price_str:
        # Loại bỏ "triệu/m²" và tính toán giá theo tỷ
        price_per_sqm = float(price_str.replace(' triệu/m²', '').replace(',', '.'))
        total_price = (price_per_sqm * area) / 1000  # Chia cho 1000 để chuyển sang tỷ
    elif 'triệu' in price_str:
        # Loại bỏ "triệu" và tính toán giá theo tỷ
        price = float(price_str.replace(' triệu', '').replace(',', '.'))
        total_price = price / 1000  # Chia cho 1000 để chuyển sang tỷ
    else:
        # Trường hợp đã là tỷ thì không cần làm gì
        total_price = float(price_str.replace(',', '.'))
    
    return total_price

# Áp dụng hàm convert_price lên từng hàng của DataFrame
df['Giá tiền'] = df.apply(convert_price, axis=1)

df.head()	
```

```python
# Kiểm tra các giá trị trong cột "Giá tiền"
print(df['Giá tiền'].unique())
```

```python
# Thay thế các giá trị NaN trong cột 'Giấy tờ pháp lý' bằng 'Không có'
df['Nội thất'].fillna('Không rõ', inplace=True)
```

```python
# Tạo cột 'Nội thất đầy đủ' dựa trên điều kiện đã cho
df['Nội thất đầy đủ'] = np.where(
    (df['Nội thất'].isin(['Không rõ', 'Không nội thất']) & (df['nội thất'] == 'không')),  # Kiểm tra điều kiện mới
    'không',  # Nếu điều kiện đúng, gán giá trị 'Không'
    np.where(
        (df['Nội thất'] != 'Not found') | (df['nội thất'] == 'có'),  # Điều kiện cho giá trị là 'Có'
        'có',  # Nếu điều kiện đúng, gán giá trị 'Có'
        np.nan  # Nếu không, gán giá trị NaN (có thể thay đổi tùy ý)
    )
)

# Xóa các cột 'Nội thất' và 'nội thất'
#df = df.drop(columns=['Nội thất', 'nội thất'])
```
